###   The top 5 player who earn highest salary

```python
player_salary= df[["short_name","wage_eur"]]
```

```python
player_salary.head()
```

```python
player_salary= player_salary.sort_values(by="wage_eur", ascending=False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(player_salary["short_name"])[0:5], list(player_salary["wage_eur"])[0:5], color=["red","pink","purple","orange","green"])
plt.show()
```

L. Messi earn highest Salary
